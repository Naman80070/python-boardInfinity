Points to be noticed -->

- P_catalog_file has a column named ID, whereas P_detail_file has a column ProductID

- P_catalog_file has a column named as Gender which has 6 unique values as shown above

Let's add a new column where we can have 3 categories such as ---
- 1) men + boys
- 2) women + girls
- 3) unisex + unisex_kids

```python
# Creating a new column for 6 unique gender types to 3 only ---

x = []
for i in P_catalog_file.Gender:
    if i == 'Men' or i == 'Boys':
        x.append('M')
    elif i == 'Women' or i=='Girls':
        x.append('F')
    else:
        x.append('U')
    
P_catalog_file['Gender_v1'] = x
```

```python
P_catalog_file.head(10)
```

```python
P_catalog_file['Gender_v1'].value_counts() # checking newly created gender values and their counts
```

```python
# Create an age_flag for adults and kids using where() of numpy library----

y = np.where((P_catalog_file['Gender'] == 'Boys')|
              (P_catalog_file['Gender'] == 'Girls')|
              (P_catalog_file['Gender'] == 'Unisex Kids'), 'kids','adult')


P_catalog_file['age_flag'] = y
```

```python
P_catalog_file.head() #let's check our new column age_flag
```

```python
P_catalog_file['age_flag'].value_counts() #here is the value count of age_flag column
```

```python
# as noticed that writing a product description does matter for selling it online ---
# hence calculating the length of the description of the product

P_catalog_file['v1'] = P_catalog_file.Description.apply(lambda x : len(x))
```
